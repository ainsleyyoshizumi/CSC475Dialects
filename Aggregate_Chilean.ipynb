```python
import os
import librosa 
import numpy as np
from glob import glob 
from tqdm import tqdm 
import pandas as pd 

y_val = []
fnames = glob ('Chilean female dataset/es_cl_female (1)/*.wav')
for fname in tqdm(fnames): 
    full_path = fname
    
    # load .wav file
    y, sr = librosa.load(full_path)
    
    # append y with fname
    y_val.append((y, fname))
    
# concatenate/append all ys with fname
concatenated_y = np.concatenate([y for y, _ in y_val])
concatenated_fname = [fname for _, fname in y_val]

print(concatenated_y)
print(concatenated_fname)
    

```

```python

# np.vstack(y_val[0])
all_data = pd.DataFrame([vals[0] for vals in y_val])
fnames = [vals[1] for vals in y_val]
all_data.index = fnames
all_data.transpose().to_csv('chilean_female_matrix.csv')
```
